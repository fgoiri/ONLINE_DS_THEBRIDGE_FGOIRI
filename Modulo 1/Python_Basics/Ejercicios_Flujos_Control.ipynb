## Ejercicio 1
Declara una variable numérica que será una hora cualquiera del día. Implementa la siguiente casuística:

* Si es entre las 0 y las 8, print "Durmiendo"
* Si es entre las 9 y las 18, print "Trabajando"
* Si es entre las 19 y las 21, print "Clase"
* Si es entre las 22 y las 24, print "Descanso"
* En cualquier otro caso, print "Transporte o error"

```python
# NO MERECE LA PENA, ES HACER LO MISMO QUE EN IF PERO MÁS LARGO!!
hora = 3

match hora:
    case hora if 0 <= hora <= 8:
        task = "Durmiendo"
    case hora if 9 <= hora <= 18:
        task = "Trabajando"
    case hora if 19 <= hora <= 21:
        task = "Clase"
    case hora if 22 <= hora <= 24:
        task = "Descanso"
    case _:
        task = "Transporte o error"

print(f"La tarea para la hora {hora} es {task}.")
```

```python
hora = 12

if 0 <= hora <= 8: # Una hora negativa caería aquí
    task = "Durmiendo"
elif 9 <= hora <= 18:
    task = "Trabajando"
elif 19 <= hora <= 21:
    task = "Clase"
elif 22 <= hora <= 24:
    task = "Descanso"
else:
    task = "Transporte o error" # Entre las horas que quedan fuera, esta en "Transporte" / Si no, habría que recoger las primeras de los elif con '<' del número final del anterior

print(f"La tarea para la hora {hora} es {task}.")
```

## Ejercicio 2
En este ejercicio vamos a implementar un calculador de precios de casas muy sencillo. Tenemos las siguientes variables:

> superficie
>
> distrito

Implementa la siguiente casuística:

1. Si el distrito es "Moncloa" o "Centro", y además la superficie es superior a 100 metros cuadrados, el precio de la casa es de 1000
2. Si el distrito es "Salamanca", y además la superficie de la casa es al menos de 150 metros, el precio de la casa es de 1500
3. Si el distrito no es "Retiro" y la superficie está entre 60 y 80 metros, el precio es de 600
4. En cualquier otro caso, el precio será de 0

```python
superficie = float(input("Por favor, introduce la superficie de la vivienda en metros cuadrados (solo el número):").replace("," , "."))
distrito = input("Genial. Ahora introduce el distrito de la vivienda:").title()

if superficie > 100 and (distrito == "Moncloa" or distrito == "Centro"):
    print(f"Para el distrito {distrito} y la superficie {superficie}, el precio de la casa es de 1000.")

elif superficie >= 150 and distrito == "Salamanca":
    print(f"Para el distrito {distrito} y la superficie {superficie}, el precio de la casa es de 1500.")

elif 60 <= superficie <= 80 and (distrito == "Moncloa" or distrito == "Centro" or distrito == "Salamanca"): # Depende de cómo entiendas el enunciado, poner != "Retiro"
    print(f"Para el distrito {distrito} y la superficie {superficie}, el precio es de 600.")

else:
    print(f"Para el distrito {distrito} y la superficie {superficie}, el precio es 0.")
```

1. Primero se compruebe si el distrito es "Retiro". Si es así, que imprima "Distrito Retiro", y si no, "Otro distrito"

## Ejercicio 6
Dada la siguiente lista:

> list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

Implementa un programa que los recorra e imprima por pantalla todos los divisibles por 5. Si nos encontramos con alguno que sea mayor que 150, detener el bucle.

```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

for i in list1:
    if i > 150:
        break
    
    if i % 5 == 0:
        print(i)
```

## Ejercicio 7
Escribe un programa en Python que imprima por pantalla todos los números que cumplan que sean divisibles por 5 y divisibles por 7, dentro del rango de valores (150, 350).

```python
numeros = list(range(150, 351))

for i in numeros:
    if (i % 5 == 0) and (i % 7 == 0):
        print(i)
```

## Ejercicio 8
Implementa un programa que imprima por pantalla el siguiente patrón

5 4 3 2 1 

4 3 2 1 

3 2 1 

2 1 

1

NOTA: NO hay lineas en blanco entre una línea y otra.

```python
# OPCIÓN 1
lista = [5, 4, 3, 2, 1] # lista = list(range(5, 0, -1)) ; print((lista))

while len(lista) > 0:
    
    print(*lista) # '*' para que no imprima los cochetes y deje espacio entre cada número
    lista.pop(0)
        
```

```python
# OPCIÓN 2: CÓMO SÍ HACERLO
lista = [5, 4, 3, 2, 1]
lista_pop = [5, 4, 3, 2, 1]

for i in lista:
    print(*lista_pop)
    lista_pop.pop(0)
```

```python
# OPCIÓN 2: CÓMO NO HACERLO
lista = [5, 4, 3, 2, 1]

for i in lista:
    print(lista)
    print(i) # El 3 pasa a estar en la posición 1 en la siguiente vuelta, por eso no coge el 4
    x = lista.pop(0)
    print(x)
```

## Ejercicio 9
En este ejercicio vamos a crear un pequeño juego. Se trata de intentar adivinar un numero del 1 al 5. Tenemos dos intentos para acertar. Pasos a seguir:

1. Ya viene implementado con la librería `random` cómo obtener un número aleatorio del 1 al 5
2. Tendrás que declarar en una variable el numero de vidas, y mediante un bucle while, comprobar que todavia quedan vidas.
3. Dentro del bucle, obtener el valor del usuario y comprobar si es ese el numero a adivinar. Si no, actualizar las vidas.
4. Si acertamos, salimos del bucle e imprimimos por pantalla "You win". Y si perdemos también salimos del bucle, pero en este caso imprimimos por pantalla "You lose".

TIP: te puede resultar útil usar la sentencia `else` cuando acabe el bucle `while`. Lo que haya dentro de ese `else` se ejecutará una vez acabe la ejecución del `while`. Lo podrás usar para cuando pierdas.

```python
import random
```

```python
vidas = 2

while vidas >= 1:
    print(f"Tienes {vidas} vidas. Puedes continuar jugando.")
    numero = random.randint(1,5) # Si quiero acertar todo el rato el mismo número, sacar esto fuera del while
    usuario = int(input("Introduce un número entero del 1 al 5:"))
    
    if numero == usuario:
        print(f"You win, el número random {numero} es igual a tu número {usuario}.")
        break
    
    elif numero != usuario:
        vidas = vidas - 1
        print(f"No has acertado, el número random {numero} no es igual a tu número {usuario}.")
    
else:
    print(f"You lose.")
```
