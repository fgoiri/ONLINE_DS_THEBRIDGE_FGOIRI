## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
''' 
Para hacer el programa orientado a objetos de la tienda, lo primero es crear la clase, en este caso Tienda. 
A continuación nos pide que las tiendas tienen 2 atributos comúnes, que son tipo: electrodoméstico, y abierta = True. 
Estos atributos se introducen justo después de crear la clase.
Después nos índica el enunciado que definamos atributos propios de cada tienda. Estos son nombre, direccion, empleados y ventas de 3 meses.
Hay que acordarse de usar bien la sintaxis.
Por otro lado, hay que definir varios métodos que nos piden. Siendo sincera, el último método he tenido que pedirle sopitas a la IA, porque
no entendía bien lo que tenía que hacer.
'''

class Tienda:
    tipo = "Electrodomésticos"
    abierta = True
    
    def __init__(self, nombre:str, direccion:str, empleados:int, venta_3_meses:list):
        
        self.nombre = nombre
        self.direccion = direccion
        self.empleados = empleados
        self.venta_3_meses = venta_3_meses
        
    def venta_total(self):
        return sum(self.venta_3_meses)
    
    def media_ventas(self):
        return (sum(self.venta_3_meses) / 3) / self.empleados # se puede hacer con la función anterior
    
    def nombre_tienda(self):
        return f"{self.nombre} - {self.direccion}"
    
    def ventas_ult_mes (self):
        return self.venta_total[-1]
    
    def proyeccion_ventas(self, x):

        factor = 1.2 if x < 1000 else 1.5
        nuevas_ventas = [venta * factor for venta in self.venta_total]
        self.venta_total = nuevas_ventas
        
        return nuevas_ventas
```

```python
''' 
Una vez definido el programa, lo probamos con las preguntas que aparecen a continuación:
'''
# 2. Crear tres tiendas con datos inventados
tienda_1 = Tienda("ElectroCool", "Avenida de Bilbao, 11", 4, [2000, 1500, 3000])
tienda_2 = Tienda("ElectroWave", "Alameda San Mamés, 24", 3, [2250, 1750, 2200])
tienda_3 = Tienda("ElectroOutlet", "Iparragirre, 34", 2, [500, 750, 1000])
# 3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
print("Tipo de tienda:",tienda_1.tipo)
print("Está la tienda abierta?:",tienda_1.abierta)
print("Nombre de la Tienda 1:",tienda_1.nombre)
print("Dirección de la Tienda 1:",tienda_1.direccion)
print("Número de empleados en la Tienda 1:", tienda_1.empleados)
print("Ventas totales en la Tienda 1:",tienda_1.venta_total())
print("Media ventas en la Tienda 1:",tienda_1.media_ventas())
print("Nombre tienda en la Tienda 1:",tienda_1.nombre_tienda())
print("Ventas en el último mes en la Tienda 1:",tienda_1.ventas_ult_mes())
print("Proyección de ventas en la Tienda 1:",tienda_1.ventas_ult_mes())
# 4. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
tiendas = [tienda_1, tienda_2, tienda_3]
for tienda in tiendas:
    print(tienda.nombre, "ventas último mes:", tienda.ventas_ult_mes())
# 5. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
for tienda in tiendas:
    if "Avenida" in tienda.direccion:
        print(tienda.nombre)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
'''
Para hacer el programa orientado a objetos de los perros, lo primero es crear la clase, en este caso Perro. 
A continuación nos dice que los atributos comúnes de los perros son: patas = 4, orejas = 2, ojos = 2 y velocidad = 0.
Estos atributos se introducen justo después de crear la clase.
Después nos índica el enunciado que definamos atributos de inicio que son: raza (que tiene que ser un string), color de inicio Marrón, y 
también es un string, y dueño ninguno, por lo que se pone igual a None.
Otra vez recordar de usar bien la sintaxis, no olvidarse de self.
Nos han pedido definir vario métodos que se realiza usando el comando def. El método parar era bastante sencillo, simplemente había que poner
la velocidad es igual a 0. Por otro lado, para el comando andar, hay que definir una nueva variable llamada aumento de la velocidad.
Por último para que el perro ladre, solo hay que hablarle, por lo que se le da un texto y te devuelve el Guau.
'''

class Perro:
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0
    
    def __init__(self, raza:str, color:str = "Marron", dueno = False):
        
        self.raza = raza
        self.color = color
        self.dueno = dueno
        
    def andar(self, aumento_velocidad):
        self.velocidad = self.velocidad + aumento_velocidad
        return self.velocidad
    
    def parar(self):
        self.velocidad = 0
        return self.velocidad
        
    def ladrar(self, texto):
        
        return "GUAU! " + texto
```

```python
''' 
Una vez definido el programa, lo probamos con las preguntas que aparecen a continuación:
'''
# Crear objeto perro sin dueño. Por definición de la clase, dueño de base está sin dueño, por lo que no tengo que hacer nada.
perro_1 = Perro("Teckel", "Jabali")
# Comprobación de que los atributos de perro_1 están bien:
print("Número de patas del Perro 1:", perro_1.patas)
print("Número de orejas del Perro 1:", perro_1.orejas)
print("Número de ojos del Perro 1:", perro_1.ojos)
print("Velocidad inicial del Perro 1:", perro_1.velocidad)
print("Raza del Perro 1:", perro_1.raza)
print("Color del Perro 1:", perro_1.color)
print("Dueño del Perro 1:", perro_1.dueno)
print("Perro ladra: ", perro_1.ladrar("hola"))
# Probar que el perro anda y comprobar su velocidad (se hace con los métodos): 
print("El perro anda?:", perro_1.andar(10))
print("Velocidad del perro:", perro_1.velocidad)
# Parar al perro (con el otro método):
print("Velocidad tras parar:", perro_1.parar())
```
