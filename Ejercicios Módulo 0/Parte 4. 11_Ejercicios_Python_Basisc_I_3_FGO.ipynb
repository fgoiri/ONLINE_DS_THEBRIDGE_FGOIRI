## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1
print(A == B)
print(A != C)
print(A >= C)
print(C <= A)
print(B != C)
````

4. not False and True and not True

En el siguiente paréntesis tenemos entonces True and True cuyo resultado es `True`. Por último tenemos el `not True` que logicamente es False. Por lo tanto, el resultado de esta operación es `False`.
3. Primero se resuelve el paréntesis, donde tenemos un `or`. El operador `or` devuelve True si hay menos uno de ellos es True. Por lo tanto, el resyktadi de esta primera operación es `True`. La siguiente operación es True and False, y el resultado de esta operación es `False`.
4. Para hacer esta operación primero hay que resolver los `not`. Primero tenemos un not False, que por tanto es True, y al final tenemos un not True, que es por tanto un False. La siguiente operación por tanto es `True and True and False` y como se ha explicado antes, el operador `and` devuelve True si los tres valores son True, como hay un False, el resultado final es `False`.
