## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
for indice, elemento in enumerate(ejer_3):
    print(indice, elemento)

def buscar_duplicados(duplicado):
    indices = []
    for indice, elemento in enumerate(ejer_3):
        if elemento.lower() == duplicado.lower():
            indices.append(indice)
    if len(indices) >= 2:
        print("La palabra", duplicado, "está duplicada en las posiciones:", indices)
    else: 
        print("La palabra", duplicado, "no está duplicada.")
    return(indices)

buscar_duplicados("binario")
print(buscar_duplicados("es"))
print(buscar_duplicados("a"))
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def DNI_correcto(DNI):
    palabra = "TRWAGMYFPDXBNJZSQVHLCKE"
    if DNI.isdigit() and len(DNI) >= 7 and len(DNI) <= 8:
        numero = int(DNI)
        letra = palabra[numero % 23]
        print("La letra es:", letra)
        return True
    else:
        return False
DNI= "55555555"
print(DNI_correcto(DNI))
`````
